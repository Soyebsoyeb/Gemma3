# SLM Training Visualization

This notebook provides interactive visualizations for monitoring SLM training progress.

```python
import sys
sys.path.append('..')

import torch
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from pathlib import Path

%matplotlib inline

# Load training logs
train_losses = torch.load('../logs/train_losses.pt')
val_losses = torch.load('../logs/val_losses.pt')
```

```python
# Plot training curves
plt.figure(figsize=(12, 6))
plt.plot(train_losses, label='Train Loss', alpha=0.7)
plt.plot(val_losses, label='Validation Loss', alpha=0.7)
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Learning rate visualization
learning_rates = torch.load('../logs/learning_rates.pt')

plt.figure(figsize=(12, 6))
plt.plot(learning_rates)
plt.xlabel('Steps')
plt.ylabel('Learning Rate')
plt.title('Learning Rate Schedule')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Generate text samples during training
def generate_samples(model, tokenizer, prompts, device):
    model.eval()
    samples = []
    for prompt in prompts:
        input_ids = torch.tensor(tokenizer.encode(prompt)).unsqueeze(0).to(device)
        with torch.no_grad():
            output = model.generate(input_ids, max_new_tokens=50, temperature=0.8)
        samples.append(tokenizer.decode(output.squeeze().tolist()))
    return samples
```
